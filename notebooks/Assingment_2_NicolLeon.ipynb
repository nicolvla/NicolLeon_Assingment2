# Importing the database "American Community Survey-2022", Census Bureau
### This database provides information about the foreign-born population in the United States.
### My goal is to gather data that allows me to understand how many foreign-born individuals live in the U.S., how many are heads of household, and—based on that group—identify the total number of female heads of household, the employment status of women over the age of 16, and the poverty levels among foreign-born female heads of household who care for children under the age of 18 and under the age of 5.

```python
import pandas as pd
```

```python
fb22 = pd.read_csv(
    "../data/FB22.csv",
)
print(fb22.head())
```

### Here I select the columns that will provide the information I’m looking for. I also specify how I want the numbers to be interpreted. I chose to format some of them as integers and percentages. I set the column with the states to be read as a string.

```python
new_columns = [
    "States",
    "Total population",
    "Households",
    "Female_householder",
    "With_own_children",
    "OCCUPATIONCivilianemployedpopulation16yearsandover",
    "Femalecivilianemployedpopulation16yearsandover",
    "Families",
    "Poverty rates for families and people for whom poverty status is determined All families",
    "Female_householder_family",
    "With_children_18",
    "With_children_5",
]

new_fb2 = fb22[new_columns]

dtypes = {
    "States":"str",
    "Total population": "Int64",
    "Households": "Int64",
    "Female_householder": "float",
    "With_own_children": "float",
    "OCCUPATIONCivilianemployedpopulation16yearsandover": "Int64",
    "Femalecivilianemployedpopulation16yearsandover": "Int64",
    "Families": "Int64",
    "Poverty rates for families and people for whom poverty status is determined All families": "float",
    "Female_householder_family": "float",
    "With_children_18": "float",
    "With_children_5": "float",
}

new_fb2 = new_fb2.astype(dtypes)

print(new_fb2.head(2))
```

### I created a new column to find out, from the total number of foreign-born heads of household, how many are women in whole numbers, since I only had the percentage. I applied the same formula to find out the number of foreign-born women who care for children under the age of 18.

```python
new_fb2 ["number_single_moms"] = new_fb2 ["Female_householder"] / 100 * new_fb2 ["Households"]
new_fb2.head(3)
```

```python
new_fb2 ["number_single_moms_w_minors"] = new_fb2["With_own_children"] / 100 * new_fb2 ["Households"]
new_fb2.head(3)
```

### I created a new column to calculate the percentage of foreign-born women over the age of 16 who are employed.

```python
new_fb2 ["percentage_female_employed"] = new_fb2 ["Femalecivilianemployedpopulation16yearsandover"] / new_fb2 ["OCCUPATIONCivilianemployedpopulation16yearsandover"] * 100
new_fb2.head(3)
```

### I created a new column to determine the percentage of foreign-born heads of household out of the total foreign-born population living in the United States.

```python
new_fb2 ["households_foreign_born"] = new_fb2["Households"] / new_fb2["Total population"] * 100
new_fb2.head(3)
```

### I applied filtering from highest to lowest for each new column I created.

```python
new_fb2_sorted1 = new_fb2.sort_values(by='Total population', ascending=False)
new_fb2_sorted1.head(5)
```

```python
new_fb2_sorted2 = new_fb2.sort_values(by='households_foreign_born', ascending=False)
new_fb2_sorted2.head(5)
```
